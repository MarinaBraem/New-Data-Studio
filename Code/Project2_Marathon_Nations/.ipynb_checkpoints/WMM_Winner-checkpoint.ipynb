```python
%matplotlib inline
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt
import numpy as np


import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
import matplotlib.pyplot as pyplot
```

```python
wmm_marathons = pd.read_excel("WMM_marathons_winners.xlsx")
wmm_marathons.head(20)
```

```python
wmm_marathons.shape
```

```python
wmm_marathons.dtypes
```

```python
New_York = wmm_marathons[wmm_marathons['marathon'] == 'New York']
New_York.head()
```

```python
New_York = New_York.set_index('date')
New_York.head()
```

```python
Boston = wmm_marathons[wmm_marathons['marathon'] == 'Boston']
Boston.head()
```

```python
Boston = Boston.set_index('date')
Boston.head()
```

```python
fig = plt.figure()
ax = fig.add_subplot(121) # hoch/runter
ax2 = fig.add_subplot(122) # seitlich


Boston.groupby('m_country')['m_winner'].count().sort_values(ascending=True).plot(title='Men', color='#81D8D0', kind='barh', x='m_winner', ax=ax, y='m_country', figsize=(15,7))
Boston.groupby('f_country')['f_winner'].count().sort_values(ascending=True).plot(title='Women', color='#5B9EFF' ,kind='barh', x='f_winner', ax=ax2, y='m_country', figsize=(15,7))
```
