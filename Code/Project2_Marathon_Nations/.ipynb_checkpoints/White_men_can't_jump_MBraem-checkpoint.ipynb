```python
%matplotlib inline
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt
import numpy as np

import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
import matplotlib.pyplot as pyplot
```

```python
import matplotlib
matplotlib.style.use('ggplot')
```

```python
wmm_marathons = pd.read_csv("All_winners_marathons.csv")
wmm_marathons
```

```python
New_York = wmm_marathons[wmm_marathons['marathon'] == 'New York']
New_York.head()#.loc[100]
```

```python
New_York.m_country.value_counts()
```

```python
New_York['nothing'] = 4
New_York.set_index("date").plot(color="black", kind="bar", width=1.0, figsize=(20,4))
```

```python
New_York.head()
```

```python
colormap = {
    'USA': '#81D8D0',
    'KEN': '#1D302F',
    'ITA': '#5B9EFF',
    'MEX': '#81D8D0',
    'ETH': '#1D302F',
    'BRA': '#FF6953',
    'RSA': '#1D302F',
    'ERI': '#1D302F',
    'MAR': '#1D302F',
    'NZL': '#FFF93C',
    'GBR': '#5B9EFF',
    'TAN': '#1D302F'
}
colors = New_York.m_country.apply(lambda country: colormap[country])

New_York['nothing'] = 4
New_York.set_index("date").plot(color=[colors], kind="bar", figsize=(20,3), width=1.0, linewidth=0, legend=False)

#plt.savefig("New_York.pdf")
```

```python
#COLORCODES

#USA
#KEN || Kenya
#ITA || Italy 
#MEX || Mexico 
#ETH || Ethopia
#RSA || South Africa
#NZL || New Zealand
#ERI || Eritrea
#MAR || Marocco
#GBR || Great Britain 
#BRA || Brasil
#TAN || Tanzania

#Continents COLOR CODE
#1D302F (Africa -------> turquoise)
#FFC153 (Asia -------> orange) 
#5B9EFF (Europa -------> blue) 
#81D8D0 (North America -------> turquoise)  
#FFF93C (Oceanien -------> yellow) 
#FF6953 (South Amercia -------> red) 
```
