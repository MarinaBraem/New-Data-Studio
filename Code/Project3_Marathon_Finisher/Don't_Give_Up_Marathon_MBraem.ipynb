```python
%matplotlib inline
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt
import numpy as np


import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
import matplotlib.pyplot as pyplot
```

```python
import matplotlib
matplotlib.style.use('ggplot')
```

```python
NYM = pd.read_csv("NY_Marathon_clean.csv")
```

```python
NYM = NYM.set_index('date')
NYM.head()
```

```python
fig, ax = plt.subplots()

NYM['total_starter'].plot(color='#81D8D0', ax=ax, figsize=(8,8), label='total starter')
NYM['total_finisher'].plot(color='black', ax=ax, figsize=(8,8), label='total finisher')
#: df.plot.area();
```

```python
fig, ax = plt.subplots()

NYM['total_starter'].plot(color='black', ax=ax, figsize=(8,8), label='total starter', kind='area')
NYM['total_finisher'].plot(color='#81D8D0', ax=ax, figsize=(8,8), label='total finisher', kind='area')
plt.savefig("New_York_Marathon_Finisher_Starter.pdf")
```

```python
fig, ax = plt.subplots()

NYM['total_starter'].plot(color='#81D8D0', ax=ax, position=0, kind='bar', figsize=(20,10), label='total starter')
NYM['total_finisher'].plot(color='black', ax=ax, position=1, kind='bar', figsize=(20,10), label='total finisher')

plt.savefig("New_York_Marathon_Finisher_Starter.pdf")
```

```python
fig, ax = plt.subplots()

NYM['total_starter'].plot(color='#81D8D0', ax=ax, kind='bar', figsize=(20,10), label='total starter')
NYM['total_finisher'].plot(color='black', ax=ax, kind='bar', figsize=(20,10), label='total finisher')

plt.savefig("New_York_Marathon_Finisher_Starter.pdf")
```

```python
top_marathons = pd.read_csv("Finisher_2009.csv")
```
